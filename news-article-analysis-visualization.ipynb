```python
import pymysql
import numpy as np
import matplotlib.pyplot as plt

# connect to database
connection = pymysql.connect(host='localhost',
                             user='root',
                             db='news_article_analysis')
cursor = connection.cursor()

years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
```

```python
count_positive_views = []
count_negative_views = []

for year in years:
    query_negative = "SELECT COUNT(*) FROM articles WHERE class = 7 AND year = "+str(year)
    cursor.execute(query_negative)
    count_negative = cursor.fetchall()
    count_negative_views.append(count_negative[0][0])
    
    query_positive = "SELECT COUNT(*) FROM articles WHERE class = 5 AND year = "+str(year)
    cursor.execute(query_positive)
    count_positive = cursor.fetchall()
    count_positive_views.append(count_positive[0][0])
    
N = len(years)

positive_views = np.asarray(count_positive_views)
negative_views = np.asarray(count_negative_views)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, positive_views, width, color='#d62728')
p2 = plt.bar(ind, negative_views, width, bottom=positive_views)

plt.ylabel('Views')
plt.title('Articles showing +ve and -ve views')
plt.xticks(ind, np.asarray(years))
plt.yticks(np.arange(0, 100, 10))
plt.legend((p1[0], p2[0]), ('Positive', 'negative'))

plt.savefig('society_views.png')
plt.show()
```

```python
count_suicides = []

for year in years:
    query_suicide = "SELECT COUNT(*) FROM articles WHERE class = 4 AND year = "+str(year)
    cursor.execute(query_suicide)
    count_suicide = cursor.fetchall()
    count_suicides.append(count_suicide[0][0])
    
N = len(years)

print(count_suicides)

years_array = np.asarray(years)
suicides = np.asarray(count_suicides)

plt.xlabel('Year')
plt.ylabel('Suicides')
plt.title('Suicides among HIV patients')
plt.xticks(years_array)
plt.yticks(np.arange(0, 20, 5))

plt.plot(years, suicides)

plt.savefig('suicides.png')
plt.show()
```
